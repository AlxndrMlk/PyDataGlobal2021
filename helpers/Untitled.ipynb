```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('ggplot')
```

```python
returns = np.random.randn(6)
```

```python
time = np.arange(1, 7)

plt.figure(figsize=(8, 5))
plt.plot(time, returns)
plt.axhline(0, ls='--', color='black')
plt.ylim(-2, 2)
plt.ylabel('After tax return')
plt.xlabel('Time (h)')
plt.title('Bitcoin returns over time')
plt.show()
```

```python
lower, upper = returns -(returns * (time - 1)), returns + returns * (time - 1)
```

```python
time = np.arange(1, 7)

plt.figure(figsize=(8, 5))
plt.plot(time, returns)
plt.axhline(0, ls='--', color='black')
plt.fill_between(time, lower, upper, alpha=.2)
plt.ylim(-8, 8)
plt.ylabel('After tax return')
plt.xlabel('Time (h)')
plt.title('Bitcoin returns over time\n+ confidence intervals')
plt.show()
```
